# Regresjonslinje

```python
import pandas as pd

# lese inn data
path = "skostr_hoyde_gr14.csv"
df = pd.read_csv(path, delimiter=";")
print(df)
sko_str = list(pd.to_numeric(df["skostr"]))
hoyde = list(pd.to_numeric(df["hoyde"]))
print(sko_str)
print(hoyde)
```

```python
# regresjonslinje
regresjon = smf.ols('skostr~hoyde', data=df).fit()
regresjon.params

#sns.regplot()
```

# plott og lagre figur

```python
# Plotte observasjoner og regresjonslinje
sns.lmplot(x = 'skostr', y = 'hoyde', data = df, ci = None)
plt.xlim(38,45); plt.ylim(160,190)  
plt.ylabel('skostr [str]'); plt.xlabel('hoyde [cm]')
plt.savefig('plot4.png',bbox_inches='tight') 
plt.show()
```
